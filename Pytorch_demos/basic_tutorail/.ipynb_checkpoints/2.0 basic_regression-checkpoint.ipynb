介绍了如何利用pytorch建立一个基本的回归神经网络

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# 建立数据
torch.manual_seed(1)
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)
y = torch.pow(x, 2) + 0.1* torch.randn(x.shape)
```

```python
# 建立预测模型
class Net(torch.nn.Module):
    def __init__(self, n_features, n_hidden, n_output):
        super(Net,self).__init__()
        self.hidden = torch.nn.Linear(n_features, n_hidden)   # hidden layer
        self.out = torch.nn.Linear(n_hidden, n_output)   # output layer
    
    def forward(self, x):
        "重写forward，以定义向前传播"
        x = F.relu(self.hidden(x))
        x = self.out(x)
        return x
```

```python
regress_net = Net(n_features=1, n_hidden=10, n_output=1)
regress_net
```

```python
# 定义优化器
optimizer = torch.optim.SGD(regress_net.parameters(), lr=0.2)
# 定义损失函数
loss_fun = torch.nn.MSELoss()
```

```python
# 训练模型
plt.ion()
for t in range(200):
    prediction = regress_net(x)
    loss = loss_fun(prediction, y)   # 注意顺序，必须先预测值，后目标值，与sklearn中的设定不一样
    
    optimizer.zero_grad()   # 梯度清零
    loss.backward()    # 反向传播
    optimizer.step()
    
    if t % 5 == 0:
        plt.cla()
        plt.scatter(x_np, y_np)
        plt.plot(x_np, prediction.data.numpy(), 'r-')
        plt.text(0.5, 0, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size': 20, 'color':  'red'})
        plt.pause(0.1)
plt.ioff()
plt.show()
```
